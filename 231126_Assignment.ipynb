## M. 데이터 병합하기

```python
# 실습을 위한 데이터 프레임 만들기 1
adfm = pd.DataFrame({"x1" : ["A","B","C"],
                    "x2": [1,2,3]})
adfm
```

```python
# 데이터 프레임 만들기 2
bdfm = pd.DataFrame({"x1" : ["A","B","D"],
                    "x3": ["T","F","T"]})
bdfm
```

```python
# 데이터 프레임 만들기 3
cdfm = pd.DataFrame({"x1" : ["B","C","D"],
                    "x2": [2,3,4]})
cdfm
```

```python
# 70. 왼쪽 열을 축으로 병합하기
pd.merge(adfm, bdfm, how='left', on='x1')
```

```python
# 71. 오른쪽 열을 축으로 병합하기
pd.merge(adfm, bdfm, how='right', on='x1')
```

```python
# 72. 공통의 값만 병합
pd.merge(adfm, bdfm, how='inner', on='x1')
```

```python
# 73. 모든 값을 병합
pd.merge(adfm, bdfm, how='outer', on='x1')
```

```python
# 74. 특정한 열을 비교하여 공통 값이 존재하는 경우만 가져오기
adfm[adfm.x1.isin(bdfm.x1)]
```

```python
# 75. 공통 값이 존재하는 경우 해당 값을 제외하고 병합하기
adfm[~adfm.x1.isin(bdfm.x1)]
```

```python
# 76. 공통의 값이 있는 것만 병합
pd.merge(adfm, cdfm)
```

```python
# 77. 모두 병합
pd.merge(adfm, cdfm, how='outer')
```

```python
# 78. 어디서 병합되었는지 표시하기
pd.merge(adfm, cdfm, how='outer', indicator=True)
```

```python
# 79. 원하는 병합 만 남기기
pd.merge(adfm, cdfm, how='outer', indicator=True).query('_merge == "left_only"')
```

```python
# 80. merge 컬럼 없애기
pd.merge(adfm, cdfm, how='outer', indicator=True).query('_merge == "left_only"').drop(columns=['_merge'])
```
